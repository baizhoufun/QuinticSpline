```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
def PolyCoefficients(x, coeffs):
    o = len(coeffs)
    y = 0
    for i in range(o):
        y += coeffs[i]*x**i
    return y
```

```python
fig = plt.figure()
ar = pd.read_csv("spline.txt",delimiter='\t',header=None).values
ar1 = pd.read_csv("scatter.txt",delimiter='\t',header=None).values
l = (ar.shape[0]+1)/4


ax = fig.add_subplot(111)
t = np.linspace(0.01, 0.99, 10)

for k in range(0,l-1,1):
    cx = ar[0+k]
    cy = ar[l+k]        
    cz = ar[l+l+k]        
    ax.plot(PolyCoefficients(t,cx),PolyCoefficients(t,cy))        

plt.scatter(ar1[:,0],ar1[:,1])    

ax.set_aspect('equal','box')
plt.show()
```

```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
 
mpl.rcParams['legend.fontsize'] = 10
 
fig = plt.figure()

ax = fig.gca(projection='3d')
for k in range(0,l-1,1):
    cx = ar[0+k]
    cy = ar[l+k]    
    cz = ar[2*l+k]    
    ax.plot(PolyCoefficients(t,cx),PolyCoefficients(t,cy),PolyCoefficients(t,cz))
    

ax.scatter(ar1[:,0],ar1[:,1],ar1[:,2])
 
plt.show()
```
